```python
import numpy as np
import matplotlib.pyplot as plt
```

### Gradient Descent in 1D

```python
def f(x):
    return np.cos(2*np.pi*x) + x**2

def df(x):
    return -2*np.pi*np.sin(2*np.pi*x) + 2*x
```

```python
x=np.linspace(-2,2,2001)

plt.plot(x,f(x),x,df(x))
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y','dy'])
plt.show()
```

```python
# random start
localmin=np.random.choice(x,1) #np.array([0])#
print(f'Start: {localmin}')

learning_rate=.01
training_epochs=100

for i in range(training_epochs):
    grad=df(localmin)
    localmin=localmin-learning_rate*grad

print(f'End: {localmin}')
```

```python
plt.plot(x,f(x),x,df(x))
plt.plot(localmin,f(localmin),'ro')
plt.plot(localmin,df(localmin),'ro')

plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df(x)','f(x) min'])
plt.title('Empirical local minimum: %s'%localmin[0])
plt.show()
```
