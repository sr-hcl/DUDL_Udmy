```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from IPython import display
display.set_matplotlib_formats('svg')
```

```python
def reg(x,y,learningRate=.05,num_epochs=500,plot_final_loss=True,plot_preds_actual=True):
    # pytorch model build
    ANNreg=nn.Sequential(
        nn.Linear(1,1), # input layer
        nn.ReLU(), # Activation function
        nn.Linear(1,1) # Output layer
    )

    print('ANN used:',ANNreg)

    # # learning rate
    # learningRate=.05

    # loss function
    lossfun=nn.MSELoss()

    # optimizer(Flavour of GD to implement)
    optimizer=torch.optim.SGD(ANNreg.parameters(),lr=learningRate)

    # # train the model
    # num_epochs=500
    losses=torch.zeros(num_epochs)

    # Train Fr
    for epochi in range(num_epochs):
        # fwd pass
        yHat=ANNreg(x)

        # compute loss
        loss=lossfun(yHat,y)
        losses[epochi]=loss

        # backprop
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    # show losses

    # manually compute losses
    # final fwd pass
    predictions=ANNreg(x)
    # print('Predictions: ',predictions)

    # final loss(MSE)
    test_loss=(predictions-y).pow(2).mean()
    # print('Losses: ',test_loss)

    if plot_final_loss==True:
        plt.plot(losses.detach(),'o',markerfacecolor='w',linewidth=.1)
        plt.plot(num_epochs,test_loss.detach(),'ro')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.title('Final Loss = %g' %test_loss.item())
        plt.show()

    if plot_preds_actual==True:
        plt.plot(x,y,'bo',label='Real data')
        plt.plot(x,predictions.detach(),'rs',label='Predictions')
        plt.title(f'prediciton-data r={np.corrcoef(y.T,predictions.detach().T)[0,1]:.2f}')
        plt.legend()
        plt.show()
    
    return predictions,losses
```

```python
def create(m,N=50,plot=True):
    x=torch.randn(N,1)
    y=m*x+torch.randn(N,1)/2

    if plot==True:
        plt.plot(x,y,'s')
        plt.title(f' y vs x. m={m}')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()
    return x,y
```

```python
m_space=np.linspace(-2,2,21)
```

```python
# for _ in range(50):
for m in m_space:
    x,y=create(m,plot=False)
    pred,losses=reg(x,y,learningRate=.05,num_epochs=500,plot_final_loss=False,plot_preds_actual=False)

    loss=(pred-y).pow(2).mean()

    fig,ax=plt.subplots(1,2,figsize=(12,4))

    ax[0].plot(losses.detach(),'o',markerfacecolor='w',linewidth=.1)
    ax[0].plot(500,loss.detach(),'rs')
    ax[0].set_title(f'FinalLoss: {loss.item()} for m:{m}')
    ax[0].set_xlabel('Epoch')
    ax[0].set_ylabel('Loss')

    ax[1].plot(x,y,'bo',label='Real Data')
    ax[1].plot(x,pred.detach(),'rs',label='Predictions')
    ax[1].set_title(f'pred-data r={np.corrcoef(y.T,pred.detach().T)[0,1]:.2f}')
    ax[1].legend()
    ax[1].set_xlabel('x')
    ax[1].set_ylabel('y')

    plt.tight_layout()
    plt.show()
    
```
