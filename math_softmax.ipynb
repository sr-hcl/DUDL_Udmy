```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
# Manually numpy
z=[1,2,3]

num=np.exp(z)
den=np.sum(num)
sigma=num/den

print(sigma)
print(np.sum(sigma))
```

```python
# random integers
z=np.random.randint(-5,high=15,size=25)
print(z)

num=np.exp(z)
den=np.sum(num)
sigma=num/den

# print(sigma)
# print(np.sum(sigma))
# plt
plt.plot(z,sigma,'ko')
plt.xlabel('Original Number (z)')
plt.ylabel('Softmaxified $\sigma$')
# plt.yscale('log')
plt.title('$\sum\sigma$ = %g' %np.sum(sigma))
plt.show()
```

```python
print(sigma)
```

### Using pytorch

```python
# more involved
# create an instance of softmax activation class
softfun=nn.Softmax(dim=0)
# then apply the data to that function
sigmaT=softfun(torch.Tensor(z))
# now the result
print(sigmaT)
```

```python
# plotting both
plt.plot(sigma,sigmaT,'ko')
plt.xlabel('Manual softmax')
plt.ylabel('Pytorch.nn Softmax')
plt.title(f'The two methods correlate at r={np.corrcoef(sigma, sigmaT)[0,1]}')
plt.show()
```
